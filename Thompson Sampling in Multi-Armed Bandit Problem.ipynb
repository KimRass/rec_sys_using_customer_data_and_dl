# Thompson Sampling Simulation

```python
gdd.download_file_from_google_drive(file_id="1CmnFuDTSAaNxbZDkaBa1OtcVmTTnTsAY", dest_path="./Datasets/Multi-Armed Bandit Simulation.csv")
sim = pd.read_csv("./Datasets/Multi-Armed Bandit Simulation.csv")
```

```python
# 각 모델에 대해서 추천한 아이템이 사용자의 클릭을 유도했는지 여부를 `0`과 `1`로 나타냅니다.
sim.head()
```

```python
arms = pd.DataFrame(np.ones((3, 2)), index=["model1", "model2", "model3"], columns=["a", "b"])

tot_clicks = 0
for n_recs in range(3000 + 1):
    if n_recs%300 == 0:
        plt.subplots()
        for name, row in arms.iterrows():
            xs = np.linspace(0, 1, 100)
            ys = stats.beta.pdf(xs, row["a"], row["b"])
            
            plt.title(f"n_recs: {n_recs}")
            plt.plot(xs, ys, label=name)
            plt.legend()
    # 각 모델에서 PDF에 기반하여 샘플링을 합니다.
    samples = {np.random.beta(row["a"], row["b"]):name for name, row in arms.iterrows()}
    # 샘플 중 가장 큰 값을 산출한 모델을 선택합니다.
    model_chosen = samples[max(samples)]
    # 사용자가 클릭했는지 여부
    click = sim.loc[n_recs, model_chosen]    
    if click:
        arms.loc[model_chosen, "a"] += 1
    else:
        arms.loc[model_chosen, "b"] += 1
    
    tot_clicks += click
```
